```python
import tensorflow as tf
import os
import cv2
import numpy as np
```

```python
model=tf.keras.models.load_model("LightRecognition.h5")
model.summary()
```

```python
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
```

```python
h,w=100,100
img=image.load_img("goforward.jpg",target_size=(100,100,1))
imag=cv2.imread("goforward.jpg")
gray=cv2.cvtColor(np.float32(img),cv2.COLOR_BGR2GRAY)
out=np.expand_dims(gray,axis=0)
final_image=out/255.0

pred=model.predict(final_image)
max_index=np.argmax(pred[0])
print(max_index)
emotions=('go', 'goForward', 'goLeft', 'stop', 'stopLeft', 'warning', 'warningLeft')
predicted_emotion=emotions[max_index]
print(pred[0])
print(predicted_emotion)

plt.imshow(cv2.cvtColor(imag,cv2.COLOR_BGR2RGB))
plt.show()
```
